```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
si = pd.read_csv('../external-data/si_2017_data.csv')
raw_si = pd.read_excel('/Users/inbio/Google Drive/work/to-sort/papers-data/Si-Jun-raw-data.xlsx', 
                        sheet_name='Data')
raw_si_select = raw_si.iloc[si['si_jun_excel_raw_number']-2].copy() # -2 because 0 indexing vs starts at 2 in excel
si['width_um'] = np.array(raw_si_select['cel width (μm)'])
si['length_um'] = np.array(raw_si_select['cel length (μm)'])
si.head()
```

```python
bas = pd.read_csv('../external-data/basan_2015_data.csv')
bas.head()
```

```python
si_nut = si.query('cm_type == 0 and useless_type == 0')
bas_nut = bas.query('cm_type == 0 and useless_type == 0')
```

```python
mlr_si_width = LinearRegression()
X_si_width = si_nut[['growth_rate_per_hr']]
y_si_width = si_nut['width_um']
mlr_si_width.fit(X_si_width, y_si_width)
print('{} + {} * x (R2={})'.format(mlr_si_width.intercept_, mlr_si_width.coef_[0], 
                                   mlr_si_width.score(X_si_width,y_si_width)))
```

```python
mlr_basan_width = LinearRegression()
X_bas_width = bas_nut[['growth_rate_per_hr']]
y_bas_width = bas_nut['avg_cell_width_um']
mlr_basan_width.fit(X_bas_width, y_bas_width)
print('{} + {} * x (R2={})'.format(mlr_basan_width.intercept_, mlr_basan_width.coef_[0], 
                                   mlr_basan_width.score(X_bas_width,y_bas_width)))
```

```python
a0 = mlr_basan_width.intercept_ - mlr_basan_width.coef_[0] * mlr_si_width.intercept_ / mlr_si_width.coef_[0]
a1 = mlr_basan_width.coef_[0] / mlr_si_width.coef_[0]
def si2bas_width(y):
    return a0 + a1*y
print(a0)
print(a1)
```

```python
plt.plot(si_nut.growth_rate_per_hr, si_nut.width_um, 'go')
plt.plot(X_si_width, mlr_si_width.predict(X_si_width), '-r')
plt.plot(bas_nut.growth_rate_per_hr, bas_nut.avg_cell_width_um, 'gs')
plt.plot(X_bas_width, mlr_basan_width.predict(X_bas_width), '-r')
plt.plot(X_si_width, si2bas_width(si_nut.width_um), 'g^')
```

```python
si_nut = si.query('cm_type == 0 and useless_type == 0')
bas_nut = bas.query('cm_type == 0 and useless_type == 0')
plt.plot(si_nut.growth_rate_per_hr, si_nut.length_um+0.4, 'go')
plt.plot(bas_nut.growth_rate_per_hr, bas_nut.avg_cell_length_um, 'gs')
plt.ylim([1,5.5])
```

```python
norm_geom_si_basan = pd.DataFrame.from_dict({'source':['basan 2015']*bas.shape[0]+['si 2017']*si.shape[0]})


norm_geom_si_basan['growth_rate'] = np.concatenate([bas[['growth_rate_per_hr']], si[['growth_rate_per_hr']]])

for s in ['nutrient_type', 'cm_type', 'useless_type']:
    norm_geom_si_basan[s] = np.concatenate([bas[[s]], si[[s]]])

norm_geom_si_basan['modulation_type'] = 'nutrient'
norm_geom_si_basan['modulation_type'][norm_geom_si_basan['cm_type'] > 0] = 'cm'
norm_geom_si_basan['modulation_type'][norm_geom_si_basan['useless_type'] > 0] = 'useless'

norm_geom_si_basan['length'] = np.concatenate([bas[['avg_cell_length_um']], si[['length_um']]+0.4])
norm_geom_si_basan['width'] = np.concatenate([bas[['avg_cell_width_um']], si2bas_width(si[['width_um']])])
norm_geom_si_basan['aspect_ratio'] = norm_geom_si_basan['length'] / norm_geom_si_basan['width']
norm_geom_si_basan['SA_V_ratio'] = 2 / (norm_geom_si_basan['width']/2) * norm_geom_si_basan['aspect_ratio']  / (norm_geom_si_basan['aspect_ratio']-1/3) 
norm_geom_si_basan['volume'] = 2 * 3.1415 * (norm_geom_si_basan['width']/2)**3 * (norm_geom_si_basan['aspect_ratio']-1/3)
```

```python
def plot_data(df, prop):

    plt.figure(figsize=(10,7))

    sns.scatterplot(data=df,
                    x='growth_rate', 
                    y=prop, 
                    hue='modulation_type',
                    style='source',
                    palette={'cm':'b', 'nutrient':'g', 'useless':'r'},
                    s=100)

    for nutrient_type, datas in df.groupby('nutrient_type'):
        data_cm = datas.query("modulation_type == 'nutrient' or modulation_type == 'cm'")
        data_cm = data_cm.sort_values(['cm_type','growth_rate'],ascending=[True,False])
        plt.plot(data_cm['growth_rate'], data_cm[prop], 'b')
        data_useless = datas.query("modulation_type == 'nutrient' or modulation_type == 'useless'")
        data_useless = data_useless.sort_values(['useless_type','growth_rate'],ascending=[True,False])
        plt.plot(data_useless['growth_rate'], data_useless[prop], 'r')

    for _,datas in df.groupby('source'):
        data_nut = datas.query("modulation_type == 'nutrient'")
        data_nut = data_nut.sort_values('growth_rate')
        plt.plot(data_nut['growth_rate'], data_nut[prop], 'g')
```

```python
plot_data(norm_geom_si_basan, 'width')
```

```python
plot_data(norm_geom_si_basan, 'length')
```

```python
plot_data(norm_geom_si_basan, 'aspect_ratio')
```

```python
plot_data(norm_geom_si_basan, 'SA_V_ratio')
```

```python
plot_data(norm_geom_si_basan, 'volume')
```
