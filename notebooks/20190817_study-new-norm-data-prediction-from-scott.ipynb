```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../results-data/normalized-geometry-data.csv')
print(data.shape)
data.head()
```

```python
comp_scott = pd.read_csv('../results-data/resXX_scott-2010-model_cell-compositions/basan_2015_si_2017_taheri_2015_modulations.csv')
comp_scott = comp_scott.iloc[:data.shape[0]]
comp_scott['model_fU_nozero'] = comp_scott['model_fU'] + 0.001
comp_scott['model_fR_fU'] = comp_scott['model_fR'] + comp_scott['model_fU']
print(comp_scott.shape)
```

```python
# check order is right...
plt.plot(comp_scott['growth_rate_per_hr'], data['growth_rate'], 'ko')
```

```python
def plot_data(df, propx, propy):

    plt.figure(figsize=(10,7))

    sns.scatterplot(data=df,
                    x=propx, 
                    y=propy, 
                    hue='modulation_type',
                    style='source',
                    palette={'cm':'b', 'nutrient':'g', 'useless':'r'},
                    s=100)

    for nutrient_type, datas in df.groupby('nutrient_type'):
        data_cm = datas.query("modulation_type == 'nutrient' or modulation_type == 'cm'")
        data_cm = data_cm.sort_values(['cm_type','growth_rate'],ascending=[True,False])
        plt.plot(data_cm[propx], data_cm[propy], 'b')
        data_useless = datas.query("modulation_type == 'nutrient' or modulation_type == 'useless'")
        data_useless = data_useless.sort_values(['useless_type','growth_rate'],ascending=[True,False])
        plt.plot(data_useless[propx], data_useless[propy], 'r')

    for _,datas in df.groupby('source'):
        data_nut = datas.query("modulation_type == 'nutrient'")
        data_nut = data_nut.sort_values('growth_rate')
        plt.plot(data_nut[propx], data_nut[propy], 'g')
```

## SA / V prediction

```python
## full multi-linear prediction
mlr = LinearRegression()
X = comp_scott[['model_fE','model_fR', 'model_fU', 'model_kn', 'model_kt', 'growth_rate_per_hr']]
y = data['SA_V_ratio']
mlr.fit(X, y)
print(mlr.score(X, y))
print(mlr.intercept_)
print(mlr.coef_)
data['SA_V_ratio_pred_full_multi_linear'] = mlr.predict(X)
plot_data(data,'SA_V_ratio', 'SA_V_ratio_pred_full_multi_linear')
```

```python
## full multi-log prediction
mlr = LinearRegression()
X = np.log(comp_scott[['model_fE', 'model_fR', 'model_fU_nozero', 'model_kn', 'model_kt', 'growth_rate_per_hr']])
y = np.log(data['SA_V_ratio'])
mlr.fit(X, y)
print(mlr.score(X, y))
print(mlr.intercept_)
print(mlr.coef_)
data['SA_V_ratio_pred_full_multi_log'] = np.exp(mlr.predict(X))
plot_data(data,'SA_V_ratio', 'SA_V_ratio_pred_full_multi_log')
```

```python
## multi-log prediction without fE
mlr = LinearRegression()
X = np.log(comp_scott[['model_fR','model_kn', 'model_kt']])
y = np.log(data['SA_V_ratio'])
mlr.fit(X, y)
print(mlr.score(X, y))
print(mlr.intercept_)
print(mlr.coef_)
data['SA_V_ratio_pred_multi_log_nofE'] = np.exp(mlr.predict(X))
plot_data(data,'SA_V_ratio', 'SA_V_ratio_pred_multi_log_nofE')
```

```python
## multi-log prediction without fR nor kt
mlr = LinearRegression()
X = np.log(comp_scott[['model_fR','model_kn']])
y = np.log(data['SA_V_ratio'])
mlr.fit(X, y)
print(mlr.score(X, y))
print(mlr.intercept_)
print(mlr.coef_)
data['SA_V_ratio_pred_multi_log_nofE_nokt'] = np.exp(mlr.predict(X))
plot_data(data,'SA_V_ratio', 'SA_V_ratio_pred_multi_log_nofE_nokt')
```
